```python
"""
DATAPLOT1
"""

import matplotlib.pyplot as plt
import numpy as np

plt.axis([0, 150, 0, 100])
plt.axis('on')
plt.grid(True)

# -------- define function y1 vs x
x = np.arange(0, 150, 1)
y1 = 10 + np.exp(.035 * x)

# -------- plot y1 vs x
plt.plot(x, y1, 'b', label='y1')

# -------- plot the legend
plt.legend(loc='upper left')

plt.show()

"""
DATAPLOT2
"""

import matplotlib.pyplot as plt
import numpy as np

plt.axis([0, 150, 0, 100])
plt.axis('on')
plt.grid(True)

# -------- define data points
x = [20, 40, 60, 80, 100, 120, 140]
y1 = [30, 50, 30, 45, 70, 43, 80]
y2 = [45, 35, 40, 60, 60, 55, 70]

# -------- plot lines with labels
plt.plot(x, y1, 'b', label='Temperature')
plt.plot(x, y2, 'r', label='Pressure')

# -------- legend
plt.legend(loc='upper left')

# -------- add markers
plt.scatter(x, y1, color='b', marker='s')
plt.scatter(x, y2, color='r', marker='*', s=50)

plt.show()

"""
DATAPLOT3
"""

import matplotlib.pyplot as plt
import numpy as np

plt.axis([0, 140, 0, 100])
plt.axis('on')
plt.grid(True)

t = [20, 40, 60, 80, 100, 120, 140]  # Time
T = [30, 33, 37.5, 44, 55, 70, 86]   # Temperature
p = [1.8,2.3,3,4,5.4,7.3,9.6]  # Pressure

# -------- Plot T vs t in red on the left vertical axis.
plt.plot(t, T, color='r', label='Temperature')
plt.xlabel('Time')
plt.ylabel('Temperature', color='r')
plt.tick_params(axis='y', labelcolor='r')
plt.legend(loc='upper left')

# -------- Plot P vs t in blue on the right vertical axis.
plt.twinx()
plt.plot(t, p, color='b', label='Pressure')
plt.ylabel('Pressure', color='b')
plt.tick_params(axis='y', labelcolor='b')
plt.legend(loc='upper right')

# -------- title the plot
plt.title('Test Results')

plt.show()

"""
DATAPLOT4
"""
import matplotlib.pyplot as plt
import numpy as np
t=[0,20,40,60,80,100,120]
T=[28,30,35,43,55,70,85]
p=[1.8,2.3,3,4,5.4,7.3,9.6]
fig, ax1 = plt.subplots() #set up a plot ax1 with subplots
plt.grid(True) #draw grid
ax1.set_xlabel('Time (hrs)') #label X axis of ax1
l1=plt.plot(t,T, 'r', label="Temperature") #plot temperature in red as curve l1
ax1.set_ylim([0,100]) #set Y asis limits of ax1
ax1.set_ylabel('Temperature (K)', color='r') #----------lately axis of ax1
ax2 = ax1.twinx() #set up ax2 as twin of ax1
l2=plt.plot(t, p, 'b', label='Pressure') #plot pressure in blue as curve l2
ax2.set_ylim([0,10]) # set Y axis lisits of ax2
ax2.set_ylabel('Pressure (psi)', color='b') #label Y axis of ax2
line1, =plt.plot([1], label="Temperature", color='r') #lines of legend
line2, =plt.plot([2], label="Pressure", color='b') #line 2 of legend
plt.legend(handles=[line1, line2], loc='upper left') #plot legend
plt.title("Test Data")
plt.show()


"""
DATAPLOT5
"""

import matplotlib.pyplot as plt
import numpy as np

plt.axis([0,140,0,100])
plt.axis('on')
plt.grid(True)

t=[20,40,60,80,100,120]  #–––time
T=[30,35,43,55,70,85]    #–––temperature
p=[2,3,4,5.3,7.3,9.6]     #–––pressure
v=[.6,.58,.54,.46,.35,.2]#–––volume

pp=[ ] #–––list for scaled pressure for plotting
for i in np.arange(0,len(p),1):
    pp.append(p[i]*10)  #–––scale p by 10

vv=[ ] #–––list for scaled volume for plotting
for i in np.arange(0,len(v),1):
    vv.append(v[i]*100) #–––scale volume by 100

plt.plot(t,T,color='r',label='Temperature',marker='o')  #–––plot temperature
plt.plot(t,pp,color='b',label='Pressure',marker='s')    #–––plot scaled pressure
plt.plot(t,vv,color='g',label='Volume',marker='d')      #–––plot scaled volume
plt.legend(loc='upper left')

for y in np.arange(0,100+1,20):  #–––plot pressure scale values
    a=y/10
    a=str(a)  #–––convert to string for plotting as text
    plt.text(142,y,a,color='b')

plt.xlabel('Time (hrs)')  #–––label axes
plt.ylabel('Temperature °K', color='r')
plt.text(151,65,'Pressure (psi)',rotation=90,color='b')

for y in np.arange(100,-1,-20):  #–––plot volume scale values
    a=y/100
    a=str(a)
    plt.text(162,y,a,color='g')
    plt.text(159,y+2,'_',color='g')

for y in np.arange(1,99,3):
    plt.text(157,y,'_',color='g')

plt.text(170,65,r'Volume (cm³)',rotation=90,color='g')  #–––label volume scale

plt.title('Compression Test Results')  #–––title
plt.show()

"""
REGRESSION1
"""
import matplotlib.pyplot as plt
import numpy as np

plt.axis([0,140,0,100])
plt.axis('on')
plt.grid(True)

t=[20,40,60,80,100,120]  #–––time
T=[30,35,43,55,70,85]    #–––temperature
p=[2,3,4,5.3,7.3,9.6]     #–––pressure
v=[.6,.58,.54,.46,.35,.2]#–––volume

pp=[ ] #–––list for scaled pressure for plotting
for i in np.arange(0,len(p),1):
    pp.append(p[i]*10)  #–––scale p by 10

vv=[ ] #–––list for scaled volume for plotting
for i in np.arange(0,len(v),1):
    vv.append(v[i]*100) #–––scale volume by 100

plt.plot(t,T,color='r',label='Temperature',marker='o')  #–––plot temperature
plt.plot(t,pp,color='b',label='Pressure',marker='s')    #–––plot scaled pressure
plt.plot(t,vv,color='g',label='Volume',marker='d')      #–––plot scaled volume
plt.legend(loc='upper left')

for y in np.arange(0,100+1,20):  #–––plot pressure scale values
    a=y/10
    a=str(a)  #–––convert to string for plotting as text
    plt.text(142,y,a,color='b')

plt.xlabel('Time (hrs)')  #–––label axes
plt.ylabel('Temperature °K', color='r')
plt.text(151,65,'Pressure (psi)',rotation=90,color='b')

for y in np.arange(100,-1,-20):  #–––plot volume scale values
    a=y/100
    a=str(a)
    plt.text(162,y,a,color='g')
    plt.text(159,y+2,'_',color='g')

for y in np.arange(1,99,3):
    plt.text(157,y,'_',color='g')

plt.text(170,65,r'Volume (cm³)',rotation=90,color='g')  #–––label volume scale

plt.title('Compression Test Results')  #–––title
plt.show()

import matplotlib.pyplot as plt
import numpy as np

plt.axis([0,140,0,100])
plt.axis('on')
plt.grid(True)

t=[20,40,60,80,100,120]  #–––time
T=[30,35,43,55,70,85]    #–––temperature
p=[2,3,4,5.3,7.3,9.6]     #–––pressure
v=[.6,.58,.54,.46,.35,.2]#–––volume

pp=[ ] #–––list for scaled pressure for plotting
for i in np.arange(0,len(p),1):
    pp.append(p[i]*10)  #–––scale p by 10

vv=[ ] #–––list for scaled volume for plotting
for i in np.arange(0,len(v),1):
    vv.append(v[i]*100) #–––scale volume by 100

plt.plot(t,T,color='r',label='Temperature',marker='o')  #–––plot temperature
plt.plot(t,pp,color='b',label='Pressure',marker='s')    #–––plot scaled pressure
plt.plot(t,vv,color='g',label='Volume',marker='d')      #–––plot scaled volume
plt.legend(loc='upper left')

for y in np.arange(0,100+1,20):  #–––plot pressure scale values
    a=y/10
    a=str(a)  #–––convert to string for plotting as text
    plt.text(142,y,a,color='b')

plt.xlabel('Time (hrs)')  #–––label axes
plt.ylabel('Temperature °K', color='r')
plt.text(151,65,'Pressure (psi)',rotation=90,color='b')

for y in np.arange(100,-1,-20):  #–––plot volume scale values
    a=y/100
    a=str(a)
    plt.text(162,y,a,color='g')
    plt.text(159,y+2,'_',color='g')

for y in np.arange(1,99,3):
    plt.text(157,y,'_',color='g')

plt.text(170,65,r'Volume (cm³)',rotation=90,color='g')  #–––label volume scale

plt.title('Compression Test Results')  #–––title

# ---------- same as DATAPLOT5 ----------
# (bagian ini diasumsikan mengandung definisi t, T, p, v dan setup plot seperti di script sebelumnya)

# ---------- straight line fit to Volume v vs t ----------
n = len(v)

c1 = np.sum(t)/n     # ——sum values of list t and divide by n, =average of t
c2 = np.sum(v)/n     # ——sum values of list v and divide by n, =average of v
a = np.multiply(v, t)  # ——multiply list v by t element by element = list a
c3 = np.sum(a)       # ——sum elements of a
a = np.multiply(t, t)  # ——multiply list t by t element by element = list a
c4 = np.sum(a)       # ——sum elements of a

A = (c3 - n*c1*c2) / (c4 - n*c1*c1)  # ——line parameters A and B
B = c2 - A*c1

for tp in np.arange(t[0], t[5], 2):  # ——plot line with scatter dots
    vp = A*tp + B
    vp = vp * 100  # ——scale vp for plotting
    plt.scatter(tp, vp, color='g', s=1)

# ---------- calculate RMS error ----------
sumee = 0
for i in range(len(t)):
    e = (v[i] - (A*t[i] + B))
    ee = e * e
    sumee = sumee + ee
rms = np.sqrt(sumee / n)

# ---------- labels ----------
plt.text(60, 28, 'v=At+B', color='g')
plt.arrow(78, 30, 6, 6, head_length=3, head_width=1.5, color='g', linewidth=.5)

vp1 = A*t[0] + B        # ——beginning v value of line
vp1 = '%.4f' % (vp1)    # ——reduce the number of decimal places
plt.text(2, 64, vp1, color='g')  # ——plot

vp2 = A*t[5] + B        # ——end v value of line
vp2 = '%.4f' % (vp2)
plt.text(122, 25, vp2, color='g')

Ap = '%.5f' % (A)
plt.text(65, 18, 'A=', color='g')
plt.text(72, 18, Ap, color='g')  # ——print value of A

Bp = '%.5f' % (B)
plt.text(65, 12, 'B=', color='g')
plt.text(73, 12, Bp, color='g')  # ——print value of B

rms = '%.3f' % (rms)
plt.text(95, 3, 'RMS error=', color='g')
plt.text(123, 3, rms, color='g')  # ——print RMS error

plt.show()





```
